```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("SYPS cohort83_4_sweep1.csv")
data.head()
```

```python
bar=data[(data.start5v==1)|(data.start5v==2)]
#bar=bar.rename(columns = {'leaverv':'whether they are leavers 0 no 1 yes'})
#print(bar)
plt.figure(figsize=(8,6))
sns.countplot(x='start5v',data=bar,hue='s5frendv')
#bar.plot.bar(stacked=True,legend=False)

#plt.xticks(np.arange(2),('worthwhile in some ways','very worthwhile','not worthwhile'))
#plt.ylabel('people')
plt.title('FRIENDS DECIDED TO STAY ON FOR 5TH vs whehter continue 5th? 1 yes 2 no\n')
plt.show()
```

```python
fig=plt.figure(figsize=(6,9))
labels = ['WAS STAYING ON RIGHT S5 ','WAS STAYING ON WRONG S5']
colors = ['lightblue','orange']
pie=pd.value_counts(data[(data.reflct5v==1)|(data.reflct5v==2)].reflct5v.values).to_frame()
pie.plot.pie(y=0,labels=labels,colors=colors,labeldistance = 1.1,autopct = '%3.1f%%',shadow = True, startangle = 10,pctdistance = 0.6)
plt.axis('equal')
plt.legend(loc='upper left', bbox_to_anchor=(-0.5, 1.5))
plt.show()
```

```python
fig=plt.figure(figsize=(6,9))
labels = ['IF LOOKED FOR FULL TIME JOB S5 yes ','IF LOOKED FOR FULL TIME JOB S5 no']
colors = ['lightblue','orange']
pie=pd.value_counts(data[(data.lookjobv==1)|(data.lookjobv==2)].lookjobv.values).to_frame()
pie.plot.pie(y=0,colors=colors,labeldistance = 1.1,autopct = '%3.1f%%',shadow = True, startangle = 10,pctdistance = 0.6)
plt.axis('equal')
plt.legend(loc='upper left', bbox_to_anchor=(-0.5, 1.5))
plt.show()
```
